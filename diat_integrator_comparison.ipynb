### A Comparison of the Velocity-Verlet and Runge-Kutta Integrators

```python
vvenpath = thisdir / 'temp_diat_harmonic_vv' / 'Energies.csv'
vv = pd.read_csv(vvenpath)
rk4enpath = thisdir / 'temp_diat_harmonic_rk4' / 'Energies.csv'
rk4 = pd.read_csv(rk4enpath)
xs = range(0, len(vv["TE"]), 1)
plt.plot(xs, vv["TE"], "r--", label = "Total Energy (Verlet)")
plt.plot(xs, rk4["TE"], "b--", label = "Total Energy (RK4)")
plt.xlabel("$N$")
plt.ylabel("Total Energy")
plt.legend()
plt.title("Total Energy vs. Iteration")
plt.savefig('VvVsRk4')
plt.show()
```
